Bootstrapping in Statistics:
Bootstrapping is a resampling technique that involves repeatedly sampling, with replacement, from a dataset to estimate the distribution of a statistic. It is often used to estimate the variability of a statistic (e.g., the mean or median) by generating multiple "bootstrap samples" and computing the statistic for each. This method allows for calculating standard errors and confidence intervals without assuming a particular distribution for the population.
Standard Error vs. Standard Deviation:
Standard Deviation (SD) measures the spread of individual data points in a sample around the sample mean. It quantifies the variability within the sample.
Standard Error (SE) quantifies the variability of a sample statistic, such as the sample mean, across multiple samples. It tells you how much the sample mean is expected to vary from the true population mean. The SE decreases as sample size increases.
Formula for SE
What Standard Error Represents: The standard error measures how far the sample mean is likely to be from the true population mean. Additionally, it can be interpreted as how much the sample means would vary from each other across many hypothetical samples.
```python
#Q4

import numpy as np
import pandas as pd
import seaborn as sns

# Load the Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Extract the 'sepal_length' column as our sample data
data = iris['sepal_length'].values

# Number of bootstrap samples
n_bootstrap = 10000

# Function to generate bootstrap samples and compute the statistic (mean in this case)
def bootstrap_mean(data, n_bootstrap):
    # Store the means of each bootstrap sample
    bootstrap_means = np.empty(n_bootstrap)
    
    # Generate n_bootstrap resamples and compute their means
    for i in range(n_bootstrap):
        # Sample with replacement from the data
        sample = np.random.choice(data, size=len(data), replace=True)
        # Calculate the mean of the sample and store it
        bootstrap_means[i] = np.mean(sample)
        #just have to change .mean to other parameters to calculate others.
    # Compute the 95% confidence interval
    lower_bound = np.percentile(bootstrap_means, 2.5)
    upper_bound = np.percentile(bootstrap_means, 97.5)
    
    return lower_bound, upper_bound

# Compute the 95% bootstrap confidence interval for the mean
ci_lower, ci_upper = bootstrap_mean(data, n_bootstrap)
print(f"95% Bootstrap Confidence Interval for Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```
